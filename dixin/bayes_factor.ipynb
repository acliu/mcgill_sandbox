# The Schwarz Criterion

The Schwarz information criterion (SIC) is a criterion for model selection among a finite set of models. It estimates the log of Bayes factor (BF).

```python
import numpy as np
from scipy import stats
from scipy.optimize import curve_fit
from scipy.stats import norm
import matplotlib.pyplot as plt
```

## 1. Define models

Suppose we have two independent complex Gaussian distributions $Z_i$ and $Z_j$. We define two models:
- real_pdf is the PDF of the real part of the product of $Z_i$ and $Z_j$, i.e., **Re**{$Z_iZ_j^*$}.
- null_pdf is the PDF of the difference between **Re**{$Z_iZ_j^*$} and **Re**{$Z_i'Z_j'^*$}.

```python
def real_pdf(x, sig):
    a = 1/(sig)
    b = (-np.abs(2*x))/(sig)
    return a*np.exp(b)

def null_pdf(x, s):
    return (1/(2*s**2))*np.exp(-2*np.abs(x)/(s))*(s+2*np.abs(x))
```

## 2. Compute the SIC

The Schwarz criterion is defined as $\mathrm{S}=\ln(\hat{L_1})-\ln(\hat{L_2})-\frac{1}{2}(d_1-d_2)\ln(n)$ where

- $\hat L_i$ = the maximized value of the likelihood function of the model $M_i$, i.e. $\hat L=p(x\mid\widehat\theta_i,M_i)$, where $\widehat\theta_i$ are the parameter values that maximize the likelihood function
- $x$ = the observed data
- $n$ = the number of data points in $x$, the number of observations, or equivalently, the sample size
- $d_i$ = the number of parameters estimated by the model.

```python
def bayes_factor_approx(data, m1, m2, num_bin='auto'):
    """
    Estimate twice the log of bayes factor for a set of data fitted with two models.
    
    Parameters
    ----------
    data : array_like
           Input data. The SIC is computed over the flattened array.
    m1, m2 : callable
           PDFs. For normal distribution PDF, use 'norm'.
    num_bin : integer
           Number of histogram bins. Default: 'auto'
    
    Returns
    -------
    s : float
           Return the Schwarz criterion.
    """

    # get normalized histogram data
    y, x = np.histogram(data, bins=num_bin, density=True)
    x = (x + np.roll(x, -1))[:-1] / 2.0  # taking the middle value between bin edges
    
    # fit the models to the histogramed data
    popt1 = curve_fit(m1, x, y, p0=np.std(data))[0]
    if m2 == 'norm':
        popt2 = norm.fit(data)
    else:
        popt2 = curve_fit(m2, x, y, p0=np.std(data))[0]
    
    n = len(data)
        
    # compute S
    bic1 = sum(np.log(m1(data, *popt1)))
    if m2 == 'norm':
        bic2 = sum(np.log(norm.pdf(data, *popt2)))
    else:
        bic2 = sum(np.log(m2(data, *popt2)))
    p = (len(popt1)-len(popt2))/np.log(n)
    
    # estimate twice the log of Bayes factor
    s = 2*(bic1-bic2-0.5*p)

    return s
```

## Example 1

We make 50 groups of arrays. There are 50 arrays of the same length in each group, and we change the length for each group. For example, all arrays in group A contain 10 data points while all arrays in group B contain 20 data points. We calculate the SIC values of each array fitting with model 1 (real_pdf) and model 2 (Gaussian), and average these SICs group by group. We plot the **averaged** SICs as a function of the length, i.e., sample siae. The plot shows that model 1 is strongly preferred than the Gaussian model when the number of data points is greater than ~100.

```python
sics = []
size = []
for i in range(50):
    b = []
    for j in range(50):
        if i == 0:
            size.append(50+50*j)
        x = np.random.normal(size=(2, size[j])).view(np.complex128)    # simulate random variables
        z = x[0, :] * x[1, :]    # compute the product
        z = z.real     # get the real part of the product
        b.append(bayes_factor_approx(z, real_pdf, 'norm'))    # compute 2ln(BF)
    sics.append(b)
    
# get the size of each dataset
size = np.asarray(size)/2

# average values of 2ln(BF) for each dataset
sics = np.mean(sics, axis=0)

# plot SICs over sample size
plt.figure(figsize=(8, 5))
plt.plot(size, sics, '.', label="2 $\ln\,{(BF)}$")
plt.hlines(10, size[0], size[-1], label="very strong evidence against $\mathcal{N}$")
plt.xlabel("sample size", fontsize=14)
plt.ylabel("2S", fontsize=14)
plt.title("The Schwarz criterion against sample size", fontsize=14)
plt.legend(loc='best', fontsize=10)
plt.show()
```

We can also plot a histogram of the data along with probability density functions predicted by two models. When the number of data points is only 50, it is hard to see which model fits better (or neither fits well).

```python
# generate dataset with 50 numbers of data points
x = np.random.normal(size=(2, 100)).view(np.complex128)
z = x[0, :] * x[1, :]
z = z.real

# fit each model to the simulated data
y, x = np.histogram(z, bins='auto', density=True)
x = (x + np.roll(x, -1))[:-1] / 2.0
sig1 = curve_fit(real_pdf, x, y, p0=np.std(z))[0]
mu = norm.fit(z)[0]
std = norm.fit(z)[1]

# plot the histogram and PDFs
xfit = np.linspace(min(z), max(z), 1000)
plt.figure(figsize=(8, 5))
y, x = np.histogram(z.real, bins='auto', density=True)
x = (x + np.roll(x, -1))[:-1] / 2.0
plt.plot(x, y, '.', label='Data')
plt.plot(xfit, real_pdf(xfit, sig1), label='Re(CNN) model')
plt.plot(xfit, norm.pdf(xfit, mu, std), label='Gaussian model')
plt.legend(loc='best', fontsize=10)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.show()
```

However, when we increase the number of data points to 500, the real_pdf model fits the data better.

```python
# generate dataset with 50 numbers of data points
x = np.random.normal(size=(2, 1000)).view(np.complex128)
z = x[0, :] * x[1, :]
z = z.real

y, x = np.histogram(z, bins='auto', density=True)
x = (x + np.roll(x, -1))[:-1] / 2.0

# fit each model to the simulated data
sig1 = curve_fit(real_pdf, x, y, p0=np.std(z))[0]
mu = norm.fit(z)[0]
std = norm.fit(z)[1]

# plot the histogram and PDFs
xfit = np.linspace(min(z), max(z), 1000)
plt.figure(figsize=(8, 5))
y, x = np.histogram(z.real, bins='auto', density=True)
x = (x + np.roll(x, -1))[:-1] / 2.0
plt.plot(x, y, '.', label='Data')
plt.plot(xfit, real_pdf(xfit, sig1), label='Re(CNN) model')
plt.plot(xfit, norm.pdf(xfit, mu, std), label='Gaussian model')
plt.legend(loc='best', fontsize=10)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.show()
```

## Example 2

Example 2 is the same as example 1 except that 1) the data in this example are the differences between the product of two complex random variables, and 2) we are fitting the null_pdf and gaussian distributions to the data. The plot below shows that null_pdf is strongly preferred than the Gaussian model when the number of data points is greater than ~200.

```python
sics = []
size = []
for i in range(50):
    b = []
    for j in range(50):
        if i == 0:
            size.append(50+50*j)
        x1 = np.random.normal(size=(2, size[j])).view(np.complex128)
        z1 = x1[0, :] * x1[1, :]    # simulate product 1

        x2 = np.random.normal(size=(2, size[j])).view(np.complex128)
        z2 = x2[0, :] * x2[1, :]    # simulate product 2

        n = z1.real-z2.real    # compute the difference between product 1 & 2
        
        b.append(bayes_factor_approx(n, null_pdf, 'norm'))    # compute 2ln(BF)
    sics.append(b)
    
size = np.asarray(size)/2    # get sample size
sics = np.mean(sics, axis=0)    # compute the mean of 2ln(BF) for each group of arrays

# plot the result
plt.figure(figsize=(8, 5))
plt.plot(size, sics, '.', label="2 $\ln\,{(BF)}$")
plt.hlines(10, size[0], size[-1], label="very strong evidence against $\mathcal{N}$")
plt.xlabel("sample size", fontsize=14)
plt.ylabel("2S", fontsize=14)
plt.title("The Schwarz criterion against sample size", fontsize=14)
plt.legend(loc='best', fontsize=10)
plt.show()
```

When we have only 200 data points, it is hard to tell which model is preferred by observing the PDF plots.

```python
# generate dataset with 50 numbers of data points
x1 = np.random.normal(size=(2, 400)).view(np.complex128)
z1 = x1[0, :] * x1[1, :]
x2 = np.random.normal(size=(2, 400)).view(np.complex128)
z2 = x2[0, :] * x2[1, :]
n = z1.real-z2.real

y, x = np.histogram(n, bins='auto', density=True)
x = (x + np.roll(x, -1))[:-1] / 2.0

# fit each model to the simulated data
sig1 = curve_fit(null_pdf, x, y, p0=np.std(n))[0]
mu = norm.fit(n)[0]
std = norm.fit(n)[1]

# plot the histogram and PDFs
xfit = np.linspace(min(n), max(n), 1000)
plt.figure(figsize=(8, 5))
y, x = np.histogram(n, bins='auto', density=True)
x = (x + np.roll(x, -1))[:-1] / 2.0
plt.plot(x, y, '.', label='Data')
plt.plot(xfit, null_pdf(xfit, sig1), label='null test result model')
plt.plot(xfit, norm.pdf(xfit, mu, std), label='Gaussian model')
plt.legend(loc='best', fontsize=10)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.show()
```

However, the difference between two models becomes obvious when we increase the number of data points to 2000. 

```python
# generate dataset with 50 numbers of data points
x1 = np.random.normal(size=(2, 4000)).view(np.complex128)
z1 = x1[0, :] * x1[1, :]
x2 = np.random.normal(size=(2, 4000)).view(np.complex128)
z2 = x2[0, :] * x2[1, :]
n = z1.real-z2.real

y, x = np.histogram(n, bins='auto', density=True)
x = (x + np.roll(x, -1))[:-1] / 2.0

# fit each model to the simulated data
sig1 = curve_fit(null_pdf, x, y, p0=np.std(n))[0]
mu = norm.fit(n)[0]
std = norm.fit(n)[1]

# plot the histogram and PDFs
xfit = np.linspace(min(n), max(n), 1000)
plt.figure(figsize=(8, 5))
y, x = np.histogram(n, bins='auto', density=True)
x = (x + np.roll(x, -1))[:-1] / 2.0
plt.plot(x, y, '.', label='Data')
plt.plot(xfit, null_pdf(xfit, sig1), label='null test result model')
plt.plot(xfit, norm.pdf(xfit, mu, std), label='Gaussian model')
plt.legend(loc='best', fontsize=10)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.show()
```
